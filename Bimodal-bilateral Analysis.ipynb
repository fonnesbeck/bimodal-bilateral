```python
def score_summary(domain, bimodal=False):
    subset = lsl_dr[(lsl_dr.domain==domain) & (lsl_dr.bimodal==bimodal)]
    subset['age_test'] = (subset.age_test/12).dropna().astype(int)
    subset.age_test[subset.age_test > 11] = 11
    subset = subset[subset.age_test>1]
    byage = subset.groupby('age_test')
    n = byage.study_id.count()
    mean = byage.score.mean()
    sd = byage.score.std()
    min = byage.score.min()
    max = byage.score.max()
    summary = pd.DataFrame({'Sample Size':n, 'Mean':mean, 
    'SD':sd, 'Min':min, 'Max':max})
    return summary[['Sample Size','Mean','SD','Min','Max']]
```

```python
receptive_summary = pd.concat([score_summary("Receptive Vocabulary", bimodal=True), 
                               score_summary("Receptive Vocabulary", bimodal=False)], keys=['bimodal','bilateral'])
receptive_summary.head()
```

```python
def plot_summary(summary_data, var='Sample Size', color='seagreen'):
    fig, axes = plt.subplots(nrows=1, ncols=2)
    var = 'Sample Size'
    for m,mode in enumerate(['bimodal', 'bilateral']):
        summary_data.ix[mode][var].plot(ax=axes[m], kind='bar', grid=False, color=color)
        axes[m].set_title(mode)
        for i,x in enumerate(summary_data.ix[mode][var]):
            axes[m].annotate('%i' % x, (i+0.5, x+1), va="bottom", ha="center")
```

```python
from pymc import *
import theano.tensor as T

# Extract data to local variables
age = np.array(dataset.age)
school = np.array(dataset.school)
study_id = np.array(dataset.study_id)
male = np.array(dataset.male)
sib = np.array(dataset.sib)
score = np.array(dataset.score)
bimodal = np.array(dataset.bimodal)

# Transform age to years and center
age_std = (dataset.age - dataset.age.mean())/12.

with Model() as model:
    
    # Imputation of primary language
    p_lang = Uniform('p_lang', 0, 1, value=0.1)
    non_english_masked = np.ma.masked_equal(dataset.non_english.fillna(-1).tolist(), -1)
    non_english = Bernoulli('non_english', p=p_lang, observed=non_english_masked)
    
    # Imputation of premature
    p_prem = Uniform('p_prem', 0, 1, value=0.1)
    prem_masked = np.ma.masked_equal(dataset.premature.fillna(-1).tolist(), -1)
    prem = Bernoulli('prem', p=p_prem, observed=prem_masked)
    
    # Imputation of age at amplification
    age_cats = len(lsl_dr.age_amp.unique())-1
    p_age_amp = Dirichlet("p_age_amp", np.ones(age_cats), value=[1./age_cats]*(age_cats-1))
    age_amp_masked = np.ma.masked_equal(dataset.age_amp.fillna(-1).tolist(), -1)
    age_amp = Categorical('age_amp', p_age_amp, observed=age_amp_masked)
    
    # Imputation of hearing loss
    hl_cats = len(lsl_dr.degree_hl.unique())-1
    p_hl = Dirichlet("p_hl", np.ones(hl_cats), value=[1./hl_cats]*(hl_cats-1))
    hl_masked = np.ma.masked_equal(dataset.degree_hl.fillna(-1).tolist(), -1)
    hl = Categorical('hl', p_hl, observed=hl_masked)
    
    # Imputation of technology category
    n_tech_cats = len(demographic.implant_category.value_counts())
    #n_tech_cats = len(tech_cats)
    p_tech = Dirichlet("p_tech", np.ones(n_tech_cats), value=[1./n_tech_cats]*(n_tech_cats-1))
    tech_masked = np.ma.masked_equal(dataset.implant_category.fillna(-1).tolist(), -1)
    tech = Categorical('tech', p_tech, observed=tech_masked)

    
    mean_score = Normal("mean_score", 0, 0.001)
    
    # Indices to school random effects
    unique_schools = np.unique(school)
    school_index = [list(unique_schools).index(s) for s in school]

    # School random effect
    sigma_school = Uniform("sigma_school", 0, 1000)
    tau_school = sigma_school**-2
    alpha_school = Normal("alpha_school", mu=0, tau=tau_school, shape=unique_schools.shape)
    
    # Indices to child random effects
    unique_child = np.unique(study_id)
    child_index = [list(unique_child).index(s) for s in study_id]
    
    # Individual random effect
    sigma_child = Uniform("sigma_child", 0, 1000)
    tau_child = sigma_child**-2    
    alpha_child = Normal("alpha_child", mu=50, tau=tau_child, shape=unique_child.shape)
    
    # Random intercepts
    intercept = mean_score + alpha_child[child_index] + alpha_school[school_index]
    
    # Age effect
    beta_age = Normal("beta_age", mu=0, tau=0.001)
    
    # Gender effect
    beta_male = Normal("beta_male", mu=0, tau=0.001)
    
    # Non-english primary language effect
    beta_lang = Normal("beta_lang", mu=0, tau=0.001)
    
    # Sibling count effect
    beta_sib = Normal("beta_sib", mu=0, tau=0.001)
    
    # Premature effect
    beta_prem = Normal("beta_prem", mu=0, tau=0.001)
    
    # Age at amplification effect
    beta_age_amp = Normal("beta_age_amp", mu=0, tau=0.001)
    
    # Degree of hearing loss effect
    beta_hearing_loss = Normal("beta_hearing_loss", mu=0, tau=0.001)
    
    # Bimodal effect (relatiev to bilateral)
    beta_bimodal = Normal("beta_bimodal", mu=0, tau=0.001)

    betas = [beta_age, beta_male, beta_lang, beta_sib, beta_prem, beta_age_amp,
        beta_hearing_loss, beta_bimodal]

    covs = [age_std, male, non_english, sib, prem, age_amp, hl, bimodal]
    
    theta = intercept + dot(np.transpose(covs), betas)
    
    sigma = Uniform("sigma", 0, 1000)
    tau = sigma**-2
    
    # Data likelihood
    score_like = Normal("score_like", mu=theta, tau=tau, observed=score)
    score_like_pred = Normal("score_like_pred", mu=theta, tau=tau)
    
    start = find_MAP()
```
