## Cleaning languge dataset

5 language measures:

- 3 versions of CELF
- PLS
    - pls_ac_rs: 	PLS: Auditory Comprehension Raw Score
    - pls_ac_ss: 	PLS: Auditory Comprehension Standard Score
    - pls_ec_rs: 	PLS: Expressive Communication Raw Score
    - pls_ec_ss: 	PLS: Expressive Communication Standard Score
    - pls_tl_rs: 	PLS: Total Language Score Standard Score Total
    - pls_tl_ss: 	PLS: Total Language Score Standard Score
- OWLS
    - age_test_owls: 	Age at time of testing (OWLS)
    - owls_lc_rs: 	OWLS: Listening Comprehension Raw Score
    - owls_lc_ss: 	OWLS: Listening Comprehension Standard Score
    - owls_oe_rs: 	OWLS: Oral Expression Raw Score
    - owls_oe_ss: 	OWLS: Oral Expression Standard Score
    - owls_oc_sss: 	OWLS: Oral Composite Sum of Listening Comprehension and Oral Expression Standard Scores
    - owls_oc_ss: 	OWLS: Oral Composite Standard Score
    - owls_wes_trs: 	OWLS: Written Expression Scale Total Raw Score
    - owls_wes_as: 	OWLS: Written Expression Scale Ability Score
    - owls_wes_ss: 	OWLS: Written Expression Scale Standard Score
    - owsl_lc: 	OWLS: Written Expression Scale Language Composite (Sum of written expression age-based standard score, listening comprehension standard score and oral expression standard score)
    - owls_lcss: 	OWLS: Language Composite Standard Score

```python
def score_summary(domain, bimodal=False):
    subset = lsl_dr[(lsl_dr.domain==domain) & (lsl_dr.bimodal==bimodal)]
    subset['age_test'] = (subset.age_test/12).dropna().astype(int)
    subset.age_test[subset.age_test > 11] = 11
    subset = subset[subset.age_test>1]
    byage = subset.groupby('age_test')
    n = byage.study_id.count()
    mean = byage.score.mean()
    sd = byage.score.std()
    min = byage.score.min()
    max = byage.score.max()
    summary = pd.DataFrame({'Sample Size':n, 'Mean':mean, 
    'SD':sd, 'Min':min, 'Max':max})
    return summary[['Sample Size','Mean','SD','Min','Max']]
```

```python
def plot_summary(summary_data, var='Sample Size', color='seagreen'):
    fig, axes = plt.subplots(nrows=1, ncols=2)
    var = 'Sample Size'
    for m,mode in enumerate(['bimodal', 'bilateral']):
        summary_data.ix[mode][var].plot(ax=axes[m], kind='bar', grid=False, color=color)
        axes[m].set_title(mode)
        for i,x in enumerate(summary_data.ix[mode][var]):
            axes[m].annotate('%i' % x, (i+0.5, x+1), va="bottom", ha="center")
```

```python
from pymc import Bernoulli, Normal, Uniform, invlogit, logit, deterministic, MCMC, Matplot
from pymc import Lambda, Dirichlet, Categorical, DiscreteUniform, Exponential, Gamma

def generate_model(dataset):
    
    # Extract data to local variables
    #age = np.array(dataset.age)
    school = np.array(dataset.school)
    study_id = np.array(dataset.study_id)
    score = np.array(dataset.score)
    bimodal = np.array(dataset.bimodal)
    
    # Dummy variables for test name: [CELF-4, CELF-P2, OWLS]
    # (PLS is baseline)
    test_name = np.array(pd.get_dummies(dataset.test_name).drop('PLS', axis=1))
    
    # Transform some data
    #age_std = (age - age.mean())/12.
    
    # Imputation of primary language
    p_lang = Uniform('p_lang', 0, 1, value=0.1)
    non_english_masked = np.ma.masked_array(dataset.non_english.values, dataset.non_english.isnull(), 
                                            fill_value=0)
    non_english = Bernoulli('non_english', p=p_lang, value=non_english_masked, observed=True)
    
    # Imputation of premature
    prem_masked = np.ma.masked_array(dataset.premature_weeks.values, dataset.premature_weeks.isnull(),
                                     fill_value=0)
    prem = DiscreteUniform('p_prem', 0, 16, value=prem_masked, observed=True)
    
    # Imputation of age of cochlear implant
    beta_ci = Uniform('beta_ci', 0, 100, value=30)
    age_ci_masked = np.ma.masked_array(dataset.age_ci.values, dataset.age_ci.isnull(), fill_value=0)
    age_ci = Exponential('age_ci', 1./beta_ci, value=age_ci_masked, observed=True)
    
    # Imputation of hearing loss
    hl_cats = 7
    p_hl = Dirichlet("p_hl", np.ones(hl_cats), value=[1./hl_cats]*(hl_cats-1))
    hl_masked = np.ma.masked_array(dataset.degree_hl.values, dataset.degree_hl.isnull(), fill_value=6)
    hl = Categorical('hl', p_hl, value=hl_masked, observed=True)
    
    # Imputation of test age
    alpha_age = Gamma('alpha_age', 1, 0.1, value=3)
    beta_age = Gamma('beta_age', 1, 0.1, value=15)
    age_test_masked = np.ma.masked_array(dataset.age_test.values, dataset.age_test.isnull(), 
                                         fill_value=49.999)
    test_age = Gamma('test_age', alpha_age, beta_age, value=age_test_masked, observed=True)
    test_age_std = Lambda('test_age_std', lambda age=test_age: (age - age.mean())/12.)

    # Test type mean: 0=receptive, 1=expressive
    test_mean_score = Normal("test_mean_score", 0, 0.001, value=[100]*2)
    test_type_rec = Normal('test_type_rec', 0, 0.001, value=[0]*3)
    test_type_exp = Normal('test_type_exp', 0, 0.001, value=[0]*3)
    @deterministic
    def mean_score(b0=test_mean_score, br=test_type_rec, be=test_type_exp):
        E = (dataset.test_type=='expressive').astype(int)
        R = (dataset.test_type=='receptive').astype(int)
        return b0[E] + R*(test_name.dot(br)) + E*(test_name.dot(be))
    
    # Indices to school random effects
    unique_schools = np.unique(school)
    school_index = [list(unique_schools).index(s) for s in school]

    # School random effect
    sigma_school = Uniform("sigma_school", 0, 1000, value=1)
    tau_school = sigma_school**-2
    alpha_school = Normal("alpha_school", mu=0, tau=tau_school, value=np.zeros(len(unique_schools)))
    
    # Indices to child random effects
    unique_child = np.unique(study_id)
    child_index = [list(unique_child).index(s) for s in study_id]
    
    # Individual random effect
    sigma_child = Uniform("sigma_child", 0, 1000, value=1)
    tau_child = sigma_child**-2    
    alpha_child = Normal("alpha_child", mu=50, tau=tau_child, value=np.zeros(len(unique_child)))
    
    @deterministic
    def intercept(a0=mean_score, a1=alpha_child, a2=alpha_school):
        """Random intercepts"""
        return a0 + a2[school_index] + a1[child_index]
    
    # Age effect
    #beta_age = Normal("beta_age", mu=0, tau=0.001, value=0)
    
    # Non-english primary language effect
    beta_lang = Normal("beta_lang", mu=0, tau=0.001, value=0)
    
    # Premature effect
    beta_prem = Normal("beta_prem", mu=0, tau=0.001, value=0)
    
    # Bimodal effect (relative to bilateral)
    beta_bimodal = Normal("beta_bimodal", mu=0, tau=0.001, value=0)
    
    # Age at test
    beta_test_age = Normal("beta_test_age", mu=0, tau=0.001, value=0)

    betas = [beta_test_age, beta_lang, beta_prem, beta_age_ci, beta_hearing_loss, beta_bimodal]

    @deterministic
    def theta(b0=intercept, b=betas, x_non_english=non_english, x_prem=prem, x_age_ci=age_ci, x_hl=hl,
              x_test_age=test_age_std):
        
        t = b0 + b[0]*x_test_age + b[1]*x_non_english + b[2]*x_prem 
        t += b[3]*x_age_ci + b[4]*x_hl + b[5]*bimodal
        
        return t
    
    sigma = Uniform("sigma", 0, 1000, value=10)
    tau = sigma**-2
    
    # Data likelihood
    score_like = Normal("score_like", mu=theta, tau=tau, value=score, observed=True)
    score_like_pred = Normal("score_like_pred", mu=theta, tau=tau, value=score)
    
    return MCMC(locals())
```

```python
from pymc import AdaptiveMetropolis
data_lang = lsl_dr[lsl_dr.domain=='Language']
Mlang = generate_model(data_lang)
Mlang.use_step_method(AdaptiveMetropolis, Mlang.betas)
```
